```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    confusion_matrix, precision_score, recall_score,
    f1_score, accuracy_score, roc_curve, auc
)
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\Rolaine Ramirez\Desktop\ml (hia-clms)\lending_club_train.csv')

print("Dataset Loaded Successfully")
print(data.head())
```

```python
label_mapping = {
    "Fully Paid": 1,      # Good Payer
    "Charged Off": 0      # Bad Payer
}

data["loan_status"] = data["loan_status"].map(label_mapping)
```

```python
X = pd.get_dummies(data.drop(columns=["loan_status"]), drop_first=True)
y = data["loan_status"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)
```

```python
model = RandomForestClassifier(n_estimators=200, random_state=42)
model.fit(X_train, y_train)

print("Model Training Completed")
```

```python
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]

fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
```

```python
# A. Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nCONFUSION MATRIX:")
print(cm)
```

```python
# B. Precision
precision = precision_score(y_test, y_pred)
print("Precision:", precision)
```

```python
# C. Recall
recall = recall_score(y_test, y_pred)
print("Recall:", recall)
```

```python
# D. F1 Score
f1 = f1_score(y_test, y_pred)
print("F1 Score:", f1)
```

```python
# E. Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# G. AUC Score
auc_score = auc(fpr, tpr)
print("AUC Score:", auc_score)
```
